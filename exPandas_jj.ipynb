### 국내 아이동 평판지수(csv)

```python
import pandas as pd

df = pd.read_csv('https://bit.ly/ds-korean-idol')
df2 = pd.read_csv('https://bit.ly/ds-korean-idol-2')
```

```python
df.head()
```

```python
df_copy = df.copy()
```

### 데이터프레임 합치기 (Row 기준으로 합치기)

```python
#row에 데디어를 합치기
# pd.concat 데이터프레임을 list로 합쳐준다.
# 기존 데이터프레임 밑 행에 붙여준다.
# axis=0 (default)
df_concat = pd.concat([df, df_copy], sort=False)
```

```python
# 데이터프레임명.reset_index() 
# 데이터프레임의 index를 재설정 - 데이터를 붙인 경우 인덱스를 재설정
df_concat.reset_index()
```

```python
# reset_index() 메서드 사용시 drop=True를 설정하지 않으면 기존 index가 남아있다.
df_concat.reset_index(drop=True)
```

### 데이터프레임 합치기 (column으로 합치기)

```python
df
```

```python
df2
```

```python
pd.concat([df, df2], axis=1)
```

```python
# 데이터의 NaN값을 생성하기 위해
df3 = df2.drop([2, 2])  #2행이 사라짐
df3
```

```python
# 두 개의 데이터프레임을 합칠 때 기준은
# dedault로 index값을 사용한다.
pd.concat([df, df3], axis=1)
```

```python
# 데이터프레임을 합치는데 사용하는 2개의 메서드
# 1. concat() 메서드 : row나 column 기준으로 단순하게 이어붙이기(기준 index)
# 2. merge() 메서드 : 특정 고유한 키(unique id)값을 기준으로 이어붙인다.
```

```python
# df, df2 '이름'을 기준으로 두 데이터프레임을 합치기
# pd.merge(left, right, on='기준 column', how='left')
```

```python
# left, right방식
df_right = df2.drop([1,3,5,7,9], axis=0)
df_right
```

```python
df_right = df_right.reset_index(drop=True)
df_right
```

```python
# 두 개의 데이터프레임을 합치는 기준 항목(열) '이름'
# 왼쪽프레임의 '이름' 항목을 기준으로 순서를 정한다. 
pd.merge(df, df_right, on='이름', how='left')
```

```python
pd.merge(df, df_right, on='이름', how='right')
```

```python
# inner, outer 방식
# inner : 두 개의 데이터프레임에 모두 키 값이 존재하는 경우만 합치기
# outer : 하나의 데이터프레임에 키값이 존재하기만 하면 모두 합치기
```

```python
df_left = df.drop([0,4], axis=0)
df_left
```

```python
# inner : 두 개의 데이터프레임에 모두 키 값이 존재하는 경우만 합치기
pd.merge(df_left, df_right, on='이름', how='inner')
```

```python
# outer : 하나의 데이터프레임에 키값이 존재하기만 하면 모두 합치기
pd.merge(df_left, df_right, on='이름', how='outer')
```

```python
df_right.columns = ['성함','연봉','가족수']
df_right
```

```python
# 왼쪽 데이터프레임의 기준은 '이름', 오른쪽 데이터프레임의 기준은 '성함'
# 그러나 두 개는 같은 항목
pd.merge(df, df_right, left_on='이름', right_on='성함', how='outer')
```
