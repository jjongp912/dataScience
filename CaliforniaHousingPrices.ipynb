```python
# 결측치 처리
median = housing['total_bedrooms'].median()
housing['total_bedrooms'].fillna(median, inplace=True)
```

```python
housing = pd.get_dummies(data=housing, columns=['ocean_proximity'])
```

```python
housing["ocean_proximity_<1H OCEAN"] = housing['ocean_proximity_<1H OCEAN'].astype('float')
housing["ocean_proximity_INLAND"] = housing['ocean_proximity_INLAND'].astype('float')
housing["ocean_proximity_ISLAND"] = housing['ocean_proximity_ISLAND'].astype('float')
housing["ocean_proximity_NEAR BAY"] = housing['ocean_proximity_NEAR BAY'].astype('float')
housing["ocean_proximity_NEAR OCEAN"] = housing['ocean_proximity_NEAR OCEAN'].astype('float')
```

```python
housing.info()
```

```python
# 데이터 나누기
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(housing.drop('median_house_value',1), housing['median_house_value'],  test_size=0.2, random_state=123)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train = x_train.reset_index(drop=True)
x_test  = x_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test  = y_test.reset_index(drop=True)
```

```python
x_test.head()
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
# model정의
# n_jobs=-1이면 CPU를 모두 사용
lin_model = LinearRegression(n_jobs = -1)
```

```python
# 모델 학습
lin_model.fit(x_train, y_train)
```

```python
lin_pred = lin_model.predict(x_test)
```

```python
lin_pred.shape, y_test.shape
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(lin_pred,y_test))
```

### KNN(K-Neighbors Classifier)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_jobs=-1)
knn_model.fit(x_train, y_train)
knn_pred = knn_model.predict(x_test)
```

```python
np.sqrt(mean_squared_error(knn_pred,y_test))
```

### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
for_model = RandomForestClassifier(criterion='entropy', n_estimators=10, n_jobs=-1)
for_model.fit(x_train, y_train)
for_pred = for_model.predict(x_test)
```

```python
np.sqrt(mean_squared_error(for_pred,y_test))
```

### 의사 결정 트리(Decision Tree)

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(x_train, y_train)
tree_pred = tree_model.predict(x_test)
```

```python
np.sqrt(mean_squared_error(tree_pred,y_test))
```

```python
colors = ['r', 'c', 'm', 'y', 'k', 'khaki', 'teal', 'orchid', 'sandybrown',
          'greenyellow', 'dodgerblue', 'deepskyblue', 'rosybrown', 'firebrick',
          'deeppink', 'crimson', 'salmon', 'darkred', 'olivedrab', 'olive', 
          'forestgreen', 'royalblue', 'indigo', 'navy', 'mediumpurple', 'chocolate',
          'gold', 'darkorange', 'seagreen', 'turquoise', 'steelblue', 'slategray', 
          'peru', 'midnightblue', 'slateblue', 'dimgray', 'cadetblue', 'tomato'
         ]
```

```python
my_predictions = {}
```

```python
def plot_predictions(name_, pred, actual):
    df = pd.DataFrame({'prediction': pred, 'actual': y_test})
    df = df.sort_values(by='actual').reset_index(drop=True)

    plt.figure(figsize=(12, 9))
    plt.scatter(df.index, df['prediction'], marker='x', color='r')
    plt.scatter(df.index, df['actual'], alpha=0.7, marker='o', color='black')
    plt.title(name_, fontsize=15)
    plt.legend(['prediction', 'actual'], fontsize=12)
    plt.show()

def mse_eval(name_, pred, actual):
    global predictions
    global colors

    plot_predictions(name_, pred, actual)

    mse = np.sqrt(mean_squared_error(pred, actual))
    my_predictions[name_] = mse

    y_value = sorted(my_predictions.items(), key=lambda x: x[1], reverse=True)
    
    df = pd.DataFrame(y_value, columns=['model', 'mse'])
    print(df)
    min_ = df['mse'].min() - 10
    max_ = df['mse'].max() + 10
    
    length = len(df)
    
    plt.figure(figsize=(10, length))
    ax = plt.subplot()
    ax.set_yticks(np.arange(len(df)))
    ax.set_yticklabels(df['model'], fontsize=15)
    bars = ax.barh(np.arange(len(df)), df['mse'])
    
    for i, v in enumerate(df['mse']):
        idx = np.random.choice(len(colors))
        bars[i].set_color(colors[idx])
        ax.text(v + 2, i, str(round(v, 3)), color='k', fontsize=15, fontweight='bold')
        
    plt.title('MSE Error', fontsize=18)
    plt.xlim(min_, max_)
    
    plt.show()

def remove_model(name_):
    global my_predictions
    try:
        del my_predictions[name_]
    except KeyError:
        return False
    return True
```

```python
mse_eval('LinearRegression', lin_pred, y_test)
mse_eval('KNeighborsClassifier', knn_pred, y_test)
mse_eval('RandomForestClassifier', for_pred, y_test)
mse_eval('DecisionTreeClassifier', tree_pred, y_test)
```
